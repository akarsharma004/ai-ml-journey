```python
# Problem: Coordinate of a cuboid such that i,j,k does not sum up to n.
# Using list comprehension.
x =1
y = 1
z = 2
n=3
print([[i,j,k] for i in range(x+1) for j in range(y+1) for k in range(z+1) if (i+j+k) != n  ])
```

```python
# Problem: Given the names and grades for each student in a class of  students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.
lst = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    lst.append([name,score])
unique_scores = sorted(set(grade for _,grade in lst))
second_lowest = unique_scores[1]
names = sorted([name for name,grade in lst if grade==second_lowest])
for name in names:
    print(name)     
```
