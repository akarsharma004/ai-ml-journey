```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(0,n-1):
        t = False
        for j in range(0,n-i-1):
            if arr[j]> arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                t = True
        if not t:
            break
    return arr
```

```python
# Selection Sort
def selection_sort(lst):
    n = len(lst)
    for i in range(0,n-1):
        k = i 
        for j in range(i+1,n):
            if lst[j] < lst[k]:
                k=j
        lst[i],lst[k] = lst[k],lst[i]
    return lst            
        
```

```python
# Insertion Sort
def insertion_sort(lst):
    n = len(lst)
    for i in range(1,n):
        k = lst[i]
        j = i-1
        while j >= 0 and k <lst[j]:
            lst[j+1] = lst[j]
            j-=1
        lst[j+1] = k
    return lst
```

```python
# Problem 58: Count negative number in matrix
# def countNegative(grid): slow method
#     count = 0
#     for i in grid:
#         for j in i:
#             if j < 0:
#                 count+=1
#     return count
def countNegative(grid): #method using binary search
    count = 0 
    for row in grid:
        left,right = 0, len(row)
        while left<right:
            mid = (left+right)//2
            if row[mid] < 0:
                right = mid
            else:
                left = mid +1
        count += (len(row) - left)
    return count                
```

```python
# Problem 59:Return smallest char that is lexicographically greater than target
def next_greatest_letter(letters, target):
    start , end = 0, len(letters)
    while start < end:
        mid =(end-start)//2
        if letters[mid] > target:
            end = mid
        else:
            start = mid +1
    return letters[start % len(letters)]
```

```python
next_greatest_letter(['c', 'f', 'j'],'c')
```

```python
# Problem 60: array with num(can be multiple), list of first index of occurrence and last of the target. If not [-1,-1]
def searchRange(nums, target):
    def first(nums, target):
        firsts = -1
        left,right = 0, len(nums) - 1
        while left <= right:
            mid = (left+right)//2
            if nums[mid] == target:
                firsts = mid
                right = mid -1
            elif nums[mid] > target:
                right = mid -1
            else:
                left = mid + 1
        return firsts
    def last(nums, target):
        lasts = -1
        left,right = 0, len(nums) - 1
        while left <= right:
            mid = (left+right)//2
            if nums[mid] == target:
                lasts = mid
                left = mid +1
            elif nums[mid] > target:
                right = mid -1
            else:
                left = mid + 1
        return lasts
    first_s = first(nums,target)
    if first_s == -1:
        return [-1,-1]
    last_s = last(nums,target)
    
    return [first_s,last_s ]

            
```

```python
searchRange([5, 7, 7, 8, 8, 10],8)
```

```python
# Problem 61: Min num in an array ascending order sorted & which rotated at a pivot
def findMin(nums):
    left, right = 0, len(nums)-1
    while left  < right:
        mid = (right + left)//2
        if nums[mid] > nums[right]:
            left = mid+1
        else:
            right = mid
    return nums[left]
```

```python
# Problem 68: large number stored in a list as from most significant to least significant. Add one to that large number and return the resulted number 
def plus_one(digits):
    str_digits = map(str,digits)
    num = int(''.join(str_digits))
    po = num + 1
    po = str(po)
    lst = []
    for i in po:
       lst.append(int(i)) 
    return lst
        
```

```python
#Problem 69: return the missing number form a lst containing number from range (0,n)
def find_missing_number(num):
    r = len(num) + 1
    for i in range(0, r):
        if i not in num:
            return i
        
```

```python
# Problem 70: Is array sorted?
def is_sorted(arr):
    n = len(arr)
    for i in range(0,n-1):
        if arr[i]>arr[i+1]:
            return False
    return True
```

```python
# Problem 71: move zeroes to end maintaining the order
def move_zeroes(nums):
    lst = nums
    c = 0
    for i in nums:
        if i == 0:
            c += 1
    for j in range(c):
        lst.remove(0)
        lst.append(0)
    return lst
```

```python
# Problem 73: count consecutive 1 in a binary array
def fmco(nums):
    c = 0
    for i in range(len(nums)-1):
        if nums[i] == 1 and nums[i+1] == 1:
            c +=1
    return c
```

```python
# Problem 75: first n rows of pascal triangle
def pascal_triangle(numrows):
    if numrows == 1:
        return [[1]]
    elif numrows == 2:
        return [[1],[1,1]]
    else:
        lst  = [[1],[1,1]]
        rows = 2
        while rows < numrows:
            l = []
            for i in range(len(lst[-1])-1):
                s = 0
                s = lst[-1][i] + lst[-1][i+1]
                l.append(s)
            lst.append([1]+l+[1])
            rows +=1
        return lst
```

```python
pascal_triangle(5)
```

```python
# Problem 76: rotate matrix by 90 degree clockwise
def rotate(matrix):
    n = len(matrix) 
    for i in range(n):
        for j in range(i,n):
            matrix[i][j],matrix[j][i] = matrix[j][i],matrix[i][j] # row become column (transpose matrix)
    for i in range(n):
        left = 0
        right = n -1
        while left<right:
            matrix[i][left],matrix[i][right] = matrix[i][right],matrix[i][left]
            left +=1
            right -=1
    return matrix
```

```python
#Problem 78: Print the element of a mxn format in a spiral order clockwise.
def spiral_order(matrix):
    m = len(matrix)
    n = len(matrix[0])
    left = 0
    right = n-1
    top = 0
    bottom = m-1
    lst=[]
    while top<=bottom and left<=right:
        # get the element from the row left to right
        for col in range(left,right+1):
            lst.append(matrix[top][col])
        top += 1
        # get the element from right most column from top to bottom
        for row in range(top,bottom+1):
            lst.append(matrix[row][right])
        right -=1
        # get the element from right to left from the bottom most row
        #check if there is a bottom row
        if top<=bottom and left<=right:
            for col in range(right,left-1,-1):
                lst.append(matrix[bottom][col])
        bottom -=1
        # get the element
        if top<=bottom and left<=right:
            for row in range(bottom,top-1,-1):
                lst.append(matrix[row][left])
        left +=1
    return lst
        
```

```python
# Problem 80: Reshape the mxn matrix to rxc if possible
def matrix_reshape(mat,r,c):
    m = len(mat)
    n = len(mat[0])
    if m*n == 0:
        return mat
    if m == r and n == c:
        return mat
    if m*n != r*c:
        return mat
    lst = []
    for i in range(m):
        for j in range(n):
            lst.append(mat[i][j])
    new_mat = []
    current = 0
    for row in range(r):
        row = []
        for col in range(c):
            row.append(lst[current])
            current +=1
        new_mat.append(row)
    return new_mat
```
