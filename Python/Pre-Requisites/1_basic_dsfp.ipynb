```python
#Practice 1:
def convert_temp(temp,unit):
    """this function to convert the temperature between celsius and fahrenheit"""
    unit.lower()
    if unit == 'f':
        return (temp-32)*5/9
    elif unit == 'c':
        return (temp *9/5 + 32)
    else: 
        print('Invalid unit')
```

```python
# Practice 2:
def is_strong_password(password):
    if len(password) < 8:
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if  not any(char.isdigit() for char in password):
        return False
    if not any(char in "!@#$%^&*()_+" for char in password):
        return False
    return True
```

```python
# Problem 2: Hollow Square
def hollow_square(n):
    
    if n==1:
        return ['*']
    elif n==2:
        return ['**','**']
    else:
        lst = []
        lst.append('*'*n)
        c = n-1
        count = 1
        while count < (c):
            lst.append('*'+' '*(c-1)+'*')
            count = count +1
    lst.append('*'*n)
    return lst
```

```python
# Problem 27: Mac difference between two consecutive number
def mcd(lst):
    if lst == []:
        return "Empty list"
    else:
        d = abs(lst[0]-lst[1])
        for i in range(1,len(lst)): 
            d1 = abs(lst[i]-lst[i-1])
            if d1 > d:
                d = d1
            else:
                continue
    return d
```

```python
# Problem 29:
def rotate_list(lst,k):
    if not lst:
        return []
    k = k % len(lst)
    for _ in range(k):
        lst_el = lst.pop()
        lst.insert(0,lst_el)
    return lst
```

```python
# Problem 34: merge dict with overlapping keys
def mod(dicts):
    if dicts == []:
        return False
    else:
        merge = {}
        for i in dicts:
            
            for k,v in i.items():
                if k in merge:
                    merge[k] += v
                else:
                    merge[k] = v
        return merge   
        
```

```python
# Problem 40: decimal to binary
def decimal_2_binary(n):
    if not (n>= -10**9 and n <=10**9):
        return False
    if n == 0:
        return ["0"]
    is_negative = False
    if n <0:
        is_negative = True
        n = abs(n)
    lst = []  
    while n>0:
        lst.append(str(n%2))
        n//=2
    binary = ''.join(lst[::-1])
    if is_negative==True:
        return '-' + binary
    else:
        return binary
        
```

```python
def binary_2_decimal(n):
    binary = 0
    num = []
    for i in n:
        num.append(int(i))
    l = len(num)
    for i in range(0,l):
        b=num[i]*(2**((l-1)-i))
        binary +=b
    return binary
        
        
```

```python
# Problem 42: GCD
def gcd(n,m):
    if n>m:    
        while n%m!=0:
            r = n%m
            n=m
            m=r
        return m
    else:
        while m%n!=0:
            r = m%n
            m=n
            n=r
        return n
            
```

```python
#Problem 51: subsequence
def is_subsequence(s,t):
    i =0
    j=0
    while i<len(s) and j<len(t):
        if s[i] == t[j]:
            j+=1
        i+=1
    return j == len(t)
```

```python
#Problem 52: substring
def is_substring(s,t):
    if s == t:
        return True
    elif s=='' or t=='':
        return True
    else:
        for i in t.split():
            if i not in s.split():
                return False
            else:
                continue
        return True
            
```

```python
# Problem 53: Length of longest string
def longest_word(s):
    word_len = {}
    for i in s.split():
        word_len[i] = len(i)
    value = 0
    for values in word_len.values():
        if values > value:
            value = values
    return value
        
```
