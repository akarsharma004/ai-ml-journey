```python
# Practice 1: Sum of digit of a number
def sum_of_digits(n):
    if n == 0:
        return 0
    last = n%10
    small_ans=sum_of_digits(n//10)
    ans = last + small_ans
    return ans
```

```python
# Practice 2: Power of a (base,exp)
def power(base,exp):
    if exp ==0:
        return 1
    small_answer = power(base,exp-1)
    answer = base*small_answer  
    return answer 
```

```python
# Practice 3: Check array is sorted using recursion
def checkHeadSorted(arr):
    # Base Case: Only one element or no element means array is sorted
    if(len(arr) == 0 or len(arr)==1): 
        return True
    # Calling the function with one less value and assuming it is sorted.
    ans = checkHeadSorted(arr[1:])
    # checking if the first element is lesser than the adjacent element
    if arr[0]<arr[1]:
        # returning the answer
        return ans
    else:
        return False
# Tail Recursion Function
def checkTailSorted(arr):
    if(len(arr) == 0 or len(arr)==1): 
        return True
    if arr[0]>=arr[1]:
        return False
    return checkTailSorted(arr[1:]) 
        
```

```python
print(checkHeadSorted([2,1,4,3,8,5]))
print(checkTailSorted([1,3,4,5]))
```

```python
# Practice 4: Sum of array elements
def head_sum_array(arr):
    # if only no element is there the sum 0
    if len(arr)==0:
        return 0
    # recursion of arr without first element
    ans = head_sum_array(arr[1:])
    # add the first element with the ans 
    return ans+arr[0]
def tail_sum_array(arr,accumulator=0):
     # no element in arr return accumulator
    if len(arr)==0:
        return accumulator
    # add the first element to accumulator
    accumulator +=arr[0]
    # Return the accumulator and arr with first element removed
    return tail_sum_array(arr[1:],accumulator)
    
```

```python
print(head_sum_array([1,3,4,5]))
print(tail_sum_array([1,3,4,5]))
```

```python
# Practice 5: Recursion in string; remove the given char from the string
def remove_char(s,ch):
    # Base Case: 
    if (len(s)==0 or s == ''):
        return s
    # Recursion call
    result = remove_char(s[1:],ch)
    # Our Work: assuming we get the result from the recursive call, and do the work on returning the ans.
    if (s[0]==ch):
        # If char at 0 position return the result
       return result
    else:
        # else add the and the result
       return s[0] + result
```

```python
# Practice 6: Returning subsequences of a string
def subsequences(s):
    # Base Case:
    if len(s) == 0:
        return ['']
    # Recursive case: Give the remaining string subsequence
    small_ans = subsequences(s[1:])
    # Our Work
    my_char = s[0]
    result = [] 
    result.extend(small_ans) # List to add all the element
    # Appending all the element with s[0], to the subsequences we get from th small ans
    for sub in small_ans:
        result.append(my_char + sub)
    # Returning the subsequences
    return result    
    
```

```python
subsequences('ab')
```

```python
# Practice 7: Print all subsequences of a given string using recursion

def print_subsequences(s, takenSoFar):
    # Base case: If the input string is empty,
    # print the result collected so far (a subsequence).
    if (s == '' or len(s) == 0):
        print(takenSoFar)
        return

    # Extract the first character of the string
    currentChar = s[0]
    # Remaining string after removing the first character
    smallInput = s[1:]

    # Recursive call 1: Include the current character in the subsequence
    print_subsequences(smallInput, takenSoFar + currentChar)

    # Recursive call 2: Exclude the current character from the subsequence
    print_subsequences(smallInput, takenSoFar)

    return

# Driver code: generate all subsequences of 'abcd'
print_subsequences('abcd', '')
```

```python
# Practice 8: Print a permutation of a string(all possible formation of string length n)
def print_permutation(s,takenSoFar):
    if len(s)==0:
        print(takenSoFar)
        return
    
    ourChar = s[0]
    smallInput = s[1:]
    
    for i in range(0,len(takenSoFar)+1):
        print_permutation(smallInput,takenSoFar[0:i]+ourChar+takenSoFar[i:])
        
    return
```

```python
print_permutation('abc','')
```

```python
#Practice 9: Return Permutations of a string
def return_permutation(s):
    # Base Case: Return the empty string when it become empty or it is empty
    if s == '':
        return ['']
    # Take the first character of the string
    currentChar = s[0]
    # Get all permutation of the remaining string
    smallPermutation = return_permutation(s[1:])
    # List to get all permutation
    allPermutation = []
    # Getting each permutation from remaining string
    for each in smallPermutation:
        # Adding the remaining char to all possible positions and append to the list
        for pos in  range(0,len(each)+1):
            allPermutation.append(each[0:pos]+currentChar+each[pos:])
    # Return all the permutation of the given string
    return allPermutation
```

```python
return_permutation('abc')
```

```python
#Practice 10: Return all the possible words you can bet by pressing a numbers on an old phone keypad. The number pressed will be passed as string.
keys = {'1':'','2':'abc','3':'def','4':'ghi','5':'jkl','6':'mno','7':'pqrs','8':'tuv','9':'wxyz'}
def returnAllNumbers(input):
    #Base Case: Return empty string in list as the 
    if (input==''):
        return ['']
    # List to get all the string
    result = []
    # Rest of number string
    rest = input[1:]
    # Recursion to get rest of code combination 
    restWords = returnAllNumbers(rest)
    # Character w.r.t the keys enter
    keyLetter = keys[input[0]]
    # Each char from the keys letter
    for myChar in keyLetter:
        # Attach th char with the rest of the code combination
        for word in restWords:
            result.append(myChar + word)
    # Return all the codes
    return result
```

```python
# Practice 11: Combination of all the letter (w.r.t to the num ex: 1 for a, 26 for z) from the given string of a number.
def get_char(value):
    # Get the letter from the value(input no.) using chr()
    if (value <=0) or (value>26):
        return ""
    return chr(97+value-1)
def return_codes(input):
    # Base Case: Check for empty string
    if input=='':
        return ['']
    # If only one digit return the single char wrt to digit
    if len(input)==1:
        singleChar = get_char(int(input))
        return [singleChar]
    # First single digit of the input  
    singleDigit = int(input[0:1])
    # First double digit of the input 
    doubleDigit = int(input[0:2])
    # Resulted List
    result = []
    # Recursive call for rest of the string without first single digit
    restWithoutFirst=return_codes(input[1:])
    # Adding the first single digit to the rest
    for each in restWithoutFirst:
        result.append(get_char(singleDigit)+each)
    # Checking for the double digit from the string
    if doubleDigit>=10 and doubleDigit<=26:
        # Recursive call getting rest of the code without first double digit
        restWithoutDouble = return_codes(input[2:])
        # Adding the first double digit to the rest 
        for each in restWithoutDouble:
            result.append(get_char(doubleDigit)+each)
    # Returning the rest
    return result
    
```

```python
# Practice 12: Print all the step during moving disc from source to destination in tower of hanoi. No larger disc on smaller one.
# To count the step.
step = 1
def towerOfHanoi(n,source,destination,auxiliary):
    # Base Case: If n = 0 no moves.
    global step
    if n == 0:
        return
    # if n =1, source to destination
    if n==1:
        print(step,source,'-->',destination)
        step+=1
        return
    # Recursive call to get n-1 disc from source to auxiliary; this will make the move of nth disc source to destination
    towerOfHanoi(n-1,source,auxiliary,destination)
    # Print the step for each movement
    print(step,source, '-->', destination)
    step +=1
    # Recursive call to get n-1 from the auxiliary to source
    towerOfHanoi(n-1,auxiliary,destination,source)
```

```python
# Problem 88: Count digits in a number
def count_digits(n):
    if (1<=n<=9): # if the last number is between the condition return 1
        return 1
    elif (n==0):
        return 1
    n = int(n/10) # Removing the last digit and making the no. smaller
    return 1 + count_digits(n) # Recursing the number till the condition satisfy
```

```python
# Problem 89: Fibonacci of n
def fibonacci(n):
    if n==0:
        return 0
    if n==1:
        return 1
    l = fibonacci(n-1)
    sl = fibonacci(n-2)
    ans = l + sl
    return ans
```

```python
# Problem 93:Element First Index
def find_first_element(arr,element):
    if (len(arr)==0):
        return -1
    if arr[0]==element:
        return 0
    recursion = find_first_element(arr[1:],element)
    if recursion == -1:
        return -1
    else: 
        return recursion + 1
```

```python
# Problem 93: Find last index
def find_last_index(arr,element):
    if (len(arr)==0):
        return -1
    if arr[-1] == element:
        return len(arr) - 1
    recursion = find_last_index(arr[:-1],element)
    return recursion
    
```

```python
# Problem 94: Return all index of the element 
def find_indices(arr,element,index):
    # if len of array become equals to the index we provide than return indices
    if len(arr)==index:
        return []
    # Recursive call going get all the indices satisfying condition
    indices = find_indices(arr,element,index + 1)
    # Checking the element in that index and if true appending it to the indices
    if arr[index] == element:
        # indices.append(index) last index at start
        indices.insert(0,index) # first index at start
    return indices
```

```python
# Problem 95.5: Binary Search Using Recursion
def binary_search_helper(arr,target,s,e):
    if s>e:
        return False
    m = s +(e-s)//2
    if arr[m] == target:
        return True
    if target > arr[m]:
        return binary_search_helper(arr,target,m+1,e)
    return binary_search_helper(arr,target,s,m-1)
        
def binary_search(arr,target):
    return binary_search_helper(arr,target,0,len(arr)-1)
```

```python
# Problem 96: Merge sort using recursion
# Main Recursive Function to pass unsorted array
def merge_sort(arr):
	s = 0
	e = len(arr)-1
	return merge_sort_helper(arr,s,e)
# Merge Sort Recursive Function Helper
def merge_sort_helper(arr,s,e):
# Base Case: when start index greater than eq to last index
	if s>=e:
		return arr
# Mid term index to divide the array
	m = s +(e-s)//2
# Left part recursive call
	merge_sort_helper(arr,s,m)
# Right part recursive call
	merge_sort_helper(arr,m+1,e)
# Function call to return the merge array
	merge(arr,s,m,e)
	return arr
def merge(arr,s,m,e):
	i = s  # Left start pointer
	j = m+1 # Right start pointer
	result = [] # Resulted list
	# To swap & sort the element of left and right array 
	while(i<=m and j<=e):
	# If left < right, add left
		if arr[i] < arr[j]:
			result.append(arr[i])
			i+=1
	# If left > right, add right
		elif arr[i] > arr[j]:
			result.append(arr[j])
			j+=1
	# If left == right, add right and left
		elif arr[i] == arr[j]:
			result.append(arr[i])
			result.append(arr[j])
			i+=1
			j+=1
	# Add remaining element of left
	while (i<=m):
		result.append(arr[i])
		i+=1
	# Add remaining element of right
	while (j<=e):
		result.append(arr[j])
		j+=1
	# Change the element in th original array
	startOfResult = 0
	startOfArr = s
	while (startOfArr <=e):
		arr[startOfArr] = result[startOfResult]
		startOfResult +=1
		startOfArr +=1
	return 
	
```

```python
# Problem 97: Quick Sort using recursion.
# Main Function for sorting
def quick_sort(arr):
    s = 0
    e = len(arr) - 1
    # Call Helper function with start and end index.
    return quick_sort_helper(arr,s,e)
# Function to do the recursive call
def quick_sort_helper(arr,s,e):
    if (s>=e):
        return arr
    # Get the pivot index
    pivotIndex = partition(arr,s,e)
    # Left side array with element <pivot
    quick_sort_helper(arr,s,pivotIndex-1)
    # Right side array with element >pivot
    quick_sort_helper(arr,pivotIndex+1,e)
    return arr
def partition(arr,s,e):
    pivot = arr[e]
    k = s
    rightPosition = s
    # Find the right position of the pivot in a sorted array
    while (k<=e-1):
        if (arr[k] < pivot):
            rightPosition+=1
        k+=1
    # Pivot Index
    pivotIndex = rightPosition
    # Swap the pivot to its right position 
    arr[pivotIndex],arr[e] = arr[e], arr[pivotIndex]
    
    # Partitioned the left (<pivot) and right(>pivot) side of array
    i = s
    j = e
    while(i<pivotIndex and j>pivotIndex):
        # if pivot is greater than left element
        if (arr[i]<pivot):
            i+=1
        # if pivot is less than right element
        elif (arr[j] >= pivot):
            j-=1
        else:
        # swap if both is false
            arr[i],arr[j] = arr[j],arr[i]
            i +=1
            j -=1
    return pivotIndex
```

```python
def quick_sort(arr):
    """Quick Sort using recursion (in-place)."""
    quick_sort_helper(arr, 0, len(arr) - 1)
    return arr

def quick_sort_helper(arr, low, high):
    if low < high:
        # Partition the array and get pivot index
        pivot_index = partition(arr, low, high)
        # Sort the left side
        quick_sort_helper(arr, low, pivot_index - 1)
        # Sort the right side
        quick_sort_helper(arr, pivot_index + 1, high)
# Shorter logic to partition the array
def partition(arr, low, high):
    # Choose pivot as last element
    pivot = arr[high]
    i = low - 1  # index for smaller elements

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    # Place pivot in correct position
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# Example usage
arr=[1, 10, 3, 16, 9, 15, 3, 6, 8, 33, 99, 101, 27]
print("Original:", arr)
quick_sort(arr)
print("Sorted:", arr)
```
